```python
import numpy as np

import matplotlib.cm as cm
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 150
```

```python
data = np.load('psmnist_results_2021_01_15_22_31_34.npz')
errs, basis_names, qs = data['errs'], data['basis_names'], data['qs']
```

```python
fig, ax = plt.subplots()
for i, basis_name in enumerate(basis_names):
    ax.errorbar(qs, 100 * np.mean(errs[i], axis=-1),
                yerr=np.array(np.percentile(errs[i], 25), np.percentile(errs[i], 75)),
                label=basis_name)
#ax.set_ylim(80, 100)
ax.set_ylabel('Accuracy')
ax.set_xlabel('Function basis size $q$')
ax.set_xscale('log')
#ax.set_yscale('log')
ax.grid()
ax.legend();
```
